```r
theta <- seq(0, 1, len = 100)
alpha <- 1
beta <- 1
n <- 10
```

```r
for (i in 1:n) {
  result <- sample(1:6, 1)
  if (result == 1) {
    alpha <- alpha + 1
  }else {
    beta <- beta + 1
  }
}
```

```r
y <- dbeta(theta, alpha, beta)
plot(theta, y)
alpha / (alpha + beta)
```
